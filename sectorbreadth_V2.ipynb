```python
import requests
import openpyxl
import xlsxwriter
import yfinance as yf
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
import datetime
import pylightxl

#Retrieve ETF holdings from SST
#xle_url = "https://www.ssga.com/us/en/intermediary/etfs/library-content/products/fund-data/etfs/us/holdings-daily-us-en-xle.xlsx"
#resp = requests.get(xle_url)

#Create workbook to store ETF dataframe
#output = open('C:/Users/jerem/Desktop/Python/sectorbreadth/etfdataraw.xlsx', 'wb')
#output.write(resp.content)
#output.close()

#Create dataframe from excel file
df = pd.read_excel('C:/Users/jerem/Desktop/Python/sectorbreadth/etfdataraw.xlsx', names = ['Company', 'Ticker'], skiprows=4, usecols='A:B', nrows=20)
```

```python
#Create dataframe of company names
a = 0
names = list()
for cos in df['Company']:
    if a < 20:
        cos = str(cos)
        names.append(cos)
        a = a + 1


#Put company names into dataframe and rename column
names = pd.DataFrame(names)
names.columns = ['Company Name']


#Create list of tickers
b = 0
tick = list()
for ticker in df['Ticker']:
    if b < 20:
        ticker = str(ticker)
        tick.append(ticker)
        b = b + 1


#Add column name to ticker dataframe
tickers = pd.DataFrame(tick)
tickers.columns = ['Tickers']


#Create list of ticker last closing prices
lastprice = list()
for ticker in tick:
    tickerinfo = yf.download(ticker, period = "ytd")
    lastprice.append(tickerinfo.iloc[-1, 3])


#Put closing prices into dataframe and rename column
lastprice = pd.DataFrame(lastprice)
lastprice.columns = ['ClosingPrice']
```

```python
#Calculate 21 day SMA
sma = list()
for ticker in tick:
    symbolinfo = yf.download(ticker, period = "ytd")
    smacalc = symbolinfo['Close'].rolling(window = 20, min_periods = 20).mean()
    sma.append((smacalc.iloc[-1]))


#Put SMAs into dataframe and rename column
sma = pd.DataFrame(sma)
sma.columns = ['20daysma']


#Combine lists into one dataframe
dftest = pd.concat ([names, lastprice, sma], axis=1)
print(dftest)
```
